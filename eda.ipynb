```python
#imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#critical dataset
critical = pd.read_csv("sample_data/critical-findings-sample-data.csv")
```

```python
critical
```

```python

# create a figure and axis 
fig, ax = plt.subplots(figsize=(15,5)) 
# count the occurrence of each class 
data = critical['Modality'].value_counts() 
# get x and y data 
modality = data.index 
frequency = data.values 
# create bar chart 
ax.bar(modality, frequency) 
# set title and labels 
ax.set_title('Modality Count Plot') 
ax.set_xlabel('Modality') 
ax.set_ylabel('Frequency')
```

```python

# create a figure and axis 
fig, ax = plt.subplots(figsize=(18,5)) 
# count the occurrence of each class 
data = critical['Category'].value_counts() 
# get x and y data 
modality = data.index 
frequency = data.values 
# create bar chart 
ax.bar(modality, frequency) 
# set title and labels 
ax.set_title('Category Count Plot') 
ax.set_xlabel('Category') 
ax.set_ylabel('Frequency')
```

```python
#non-critical dataset
non_critical = pd.read_csv("sample_data/non-critical-findings-sample-data.csv")
non_critical
```

```python
# create a figure and axis 
fig, ax = plt.subplots(figsize=(15,5)) 
# count the occurrence of each class 
data = non_critical['Modality'].value_counts() 
# get x and y data 
modality = data.index 
frequency = data.values 
# create bar chart 
ax.bar(modality, frequency) 
# set title and labels 
ax.set_title('Modality Count Plot') 
ax.set_xlabel('Modality') 
ax.set_ylabel('Frequency')
```

```python
import nltk
from nltk.tokenize import word_tokenize
reviews = critical.Data.str.cat(sep=' ')
#function to split text into word
tokens = word_tokenize(reviews)
vocabulary = set(tokens)
frequency_dist = nltk.FreqDist(tokens)
```

```python
# Frequency Distribution Plot
import matplotlib.pyplot as plt
frequency_dist.plot(30,cumulative=False)
plt.show()
```

```python
non_critical['Critical_Finding'] = non_critical['Critical_Finding'].replace(regex='None', value='No Critical Finding')
#replacing None with No critical Finding
non_critical['Category'] = 'No Category'
#replacing NaN with No Category
non_critical
```

```python
findings = pd.concat([critical,non_critical])
findings=findings.reset_index(drop=True) #restting index after merge
findings
```

```python
# create a figure and axis 
fig, ax = plt.subplots(figsize=(18,5)) 
# count the occurrence of each class 
data = findings['Critical_Finding'].value_counts() 
# get x and y data 
modality = data.index 
frequency = data.values 
# create bar chart 
ax.bar(modality, frequency) 
# set title and labels 
ax.set_title('Critical Finding Count Plot') 
ax.set_xlabel('Critical Finding') 
ax.set_ylabel('Frequency')
```
