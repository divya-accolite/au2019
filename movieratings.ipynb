```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('movies.dat' , sep="::",names=['movieid','title','genre'],header = None)
print(movies.head())
ratings = pd.read_csv('ratings.dat' , sep="::",names=['userid','movieid','rating','timestamp'],header = None)
print(ratings.head())
users = pd.read_csv('users.dat' , sep="::",names=['userid','gender','age','occupation','zipcode'],header = None)
print(users.head())
```

```python
sub_df = pd.merge(movies[['movieid','title']],ratings[['userid','movieid','rating']],on='movieid')
sub_df
```

```python
master_data = pd.merge(sub_df,users[['userid','gender','age','occupation']],on='userid')
master_data
```

```python
master_data.isnull().any().any() #checks whether there are any null values 
#since it returned false , there are no null values
```

```python

genres = set()
for g in movies['genre'].str.split('|'):
    genres = genres.union(set(g))
    
genres

#taking out all the genre and putting them in a set removes duplicate elements
# and retains the unique elements
```

```python
movies = movies['genre'].str.get_dummies('|')
print (movies)
#splits the genre based on the pipe separator 
# and get_dummies assigns one-hot encoding
```

```python
master_data.head()
```

```python
print(movies)
print(ratings)
print(users)
```

```python
final_data = pd.merge(movies,ratings,on='movieid')
final_data = pd.merge(final_data,users,on='userid')
final_data
```

```python
final_data = pd.concat([final_data.drop('genre', axis=1), final_data['genre'].str.get_dummies('|')], axis=1)
final_data
#one hot encoding on the genre column
```

```python
final_data = pd.concat([final_data.drop('gender', axis=1), final_data['gender'].str.get_dummies()], axis=1)
final_data
#applying one hot encoding for the gender column
```

```python
plt.figure(figsize=(25,10))
sns.heatmap(final_data.corr(), annot=True, cmap=plt.cm.Reds)
plt.show()
#plotting correlation map
```

```python
#Correlation with output variable
cor=final_data.corr()
cor_target = abs(cor["rating"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.5]
relevant_features
#there are no varaiables with accuracy greater than 0.5
#thus the ability of these variables to predict rating will be low
```

```python
feature = final_data[final_data.columns[~final_data.columns.isin(["title","rating","zipcode"])]]
target = final_data.rating

from sklearn.model_selection import train_test_split
#splitting dataset into trainiing and testing
feature_train, feature_test, target_train, target_test = train_test_split(feature, target, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(feature_train,target_train)
```

```python
target_pred = lr.predict(feature_test)
```

```python
i=0
for elmt in target_pred:
    target_pred[i] = int(round(elmt))
    i+=1
    
target_pred = target_pred.astype(int)      
df = pd.DataFrame({'Actual': target_test, 'Predicted': target_pred})
df1 = df.head(25)
df1
```

```python
plt.scatter(target_test,target_pred) 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(target_test, target_pred)  
```
